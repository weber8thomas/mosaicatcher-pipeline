```r
# Load the data
load("/scratch/tweber/DATA/MC_DATA/PAPER_RUNS/TMP/scNOVA/heatmap_data_and_breaks.RData")

# Limit to 30 genes
heatmap_data <- heatmap_data
#heatmap_data = heatmap_data[rownames(heatmap_data) != "BM510x3PE20404.bam", ]
rownames(heatmap_data) <- sub("\\.bam$", "", rownames(heatmap_data))



# Define the annotation colors
anno_colors <- list(Subclonality = c(clone1 = "blue", clone2 = "blue"))

# Define the color palette for the heatmap
breaksList <- seq(-1.5, 1.5, by = 0.1)
breaksList <- append(breaksList, 2)
breaksList <- append(breaksList, -2, 0)
mycol <- colorRampPalette(c("blue", "white", "red"))(length(breaksList) - 1)


options(repr.plot.width = 25, repr.plot.height = 15, repr.plot.res = 100)


rownames(merged_data) <- merged_data$Row.names


# Convert to factor to ensure the levels and their order
annotation_df <- data.frame(Subclonality = factor(merged_data$Subclonality, levels = c("clone1", "clone2")))
rownames(annotation_df) = rownames(merged_data)

# Create a named vector for the color mapping
anno_colors <- c(clone1 = "green", clone2 = "orange") 



# Run pheatmap and save the result
pheatmap_result <- pheatmap(
    heatmap_data, 
    annotation_row = annotation_df, 
    show_rownames = F, 
    show_colnames = T, 
    cluster_cols = T, 
    cluster_rows = T,         
    scale = "column", 
    col = mycol, 
    breaks = breaksList, 
    clustering_distance_rows = "euclidean", 
    clustering_method = "ward.D"
)


# Extract row clustering
tree <- pheatmap_result$tree_col


# Extract clusters at top level
cut_clusters <- cutree(tree, k = 2)  # Assuming you have 2 top-level clusters; adjust if needed.

# Extract the first 20 genes for each top-level cluster
cluster1_genes <- names(cut_clusters[cut_clusters == 1])[1:50]
cluster2_genes <- names(cut_clusters[cut_clusters == 2])[1:50]

# Combine genes from all clusters to get the top genes to be displayed
top_genes <- c(cluster1_genes, cluster2_genes)

# Subset heatmap_data for the top genes
heatmap_subset <- heatmap_data[, top_genes]


# Subset subclonality data to match top genes
subclonality_subset <- heatmap_data[, top_genes, drop = FALSE]






# Replot the heatmap with the subset of genes and the annotation
pheatmap(
    heatmap_subset, 
    annotation_row = annotation_df, 
    show_rownames = TRUE, 
    show_colnames = T, 
    cluster_cols = F, 
    cluster_rows = F, 
    scale = "column", 
    col = mycol, 
    breaks = breaksList, 
    clustering_distance_rows = "euclidean", 
    clustering_method = "ward.D", 
#    annotation_colors = list(Subclonality = anno_colors)
)

```
